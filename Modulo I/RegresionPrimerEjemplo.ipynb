### Otra forma de accesar los datos

## Regresión Lineal Univariada

$$y= \theta_0 +\theta_1* x_1$$
$$salario = \theta_0+\theta_1 * experiencia$$

```
#importar las librerias
import numpy as np
import matplotlib.pyplot as plt

# separar los datos
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values

# Separa el conjunto de datos en entrenamiento y pruebas

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state = 0)
print(X.shape)
print(X_train.shape)
```

```
print(X_train.shape)
```

### Aprendizaje a Partir de Datos (Scikit Learn)

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
print(regressor.intercept_)
print(regressor.coef_)
```

### Predicción de resultados conjunto de prueba

```
y_pred = regressor.predict(X_test)
```

```
print(y_pred)
```

### Visualización de resultados conjunto de entrenamiento

```
plt.scatter(X_train, y_train, color= 'red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salario vs Experiencia (Conjunto Entrenamiento)')
plt.xlabel('Años de Experiencia')
plt.ylabel('Salario')
plt.show()
```
